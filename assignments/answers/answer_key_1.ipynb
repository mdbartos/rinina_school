### 6) Comparing word frequencies in project gutenberg

The following class will 'read' a book from project gutenberg and return the frequency of each word used in the text.

```python
import urllib2 as url
import pandas as pd
import numpy as np
import string


class read_gutenberg():
    def __init__(self):
        self.collection = {}

    def read_book(self, urltxt, author, title, year=None, gender=None, nation=None, **kwargs):
        
        meta_d = {'year': year, 'gender': gender, 'nation': nation}
        meta_d.update(kwargs)
        
        f = url.urlopen(urltxt)
        r = ''.join(f.readlines()).split('***')[2].lower()
        
        for x in [i for i in string.punctuation.replace("'", '')]:
            r = r.replace(x, ' ')

        s = pd.Series(r.split())
        df = pd.DataFrame()
        df[title] = s
        g_words = df.groupby(title).size()

        self.collection.update({author: {}})
        self.collection[author].update({title: {}})
        self.collection[author][title].update({'words': g_words})
        self.collection[author][title].update({'meta': meta_d})
```

Initiate the class to a variable like so:

```python
my_books = read_gutenberg()
```
