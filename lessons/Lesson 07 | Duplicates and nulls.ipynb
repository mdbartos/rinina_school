# 1. Dealing with missing data

## 1.1. Dropping nulls in a Series

Our DataFrame lists the addresses of every property in Maricopa county in the <b>'SITE_ADDR'</b> column.

When we examine the addresses listed our database, we find that there are a lot of missing entries.

```python
df_subset['SITE_ADDR']
```

We can drop all the missing entries using the <b>.dropna()</b> method.

```python
df_subset['SITE_ADDR'].dropna()
```

## 1.2. Dropping nulls across multiple rows/columns

### 1.2.1. Drop columns containing nulls

We can drop <b>columns</b> containing nulls by supplying the <b>axis=1</b> argument to the <b>.dropna()</b> method.

By default, this method drops columns containing ANY nulls.

```python
df_subset.dropna(axis=1)
```

This operation shows us that the 'PROP_TYPE' and 'LAND_ASSDV' columns contain no nulls.

### 1.2.2. Drop rows containing nulls

We can drop rows containing nulls by supplying the <b>axis=0</b> argument to the <b>.dropna()</b> method.

By default, this method drops rows containing ANY nulls.

```python
df_subset.dropna(axis=0)
```

### 1.2.3. Dropna logic: Any, all, and subset

#### 1.2.3.1. Any

The default logic for the <b>.dropna()</b> method is to drop rows/columns containing ANY nulls.

We can specify the logic for the <b>.dropna()</b> method using the <b>'how'</b> keyword argument.

The following example (equivalent to the default behavior) will drop rows containing ANY nulls in the four selected columns.

```python
df_subset[['SITE_ADDR', 'S_CITY', 'S_ZIP', 'SALE_PRICE']].dropna(axis=0, how='any')
```

#### 1.2.3.2. All

Sometimes we may want to drop rows or columns only when the entire row/column is null.

To do this we can supply the <b>how='all'</b> argument to the <b>dropna()</b> method.

In the following example, we drop rows containing ALL nulls for the four selected columns.

```python
df_subset[['SITE_ADDR', 'S_CITY', 'S_ZIP', 'SALE_PRICE']].dropna(axis=0, how='all')
```

#### 1.2.3.3. Subset

Sometimes we may want to drop rows or columns only when a subset of the columns contain nulls.

To do this we can supply the <b>subset</b> argument to the <b>dropna()</b> method.

In the following example, we drop rows containing ANY nulls in the 'S_CITY' and 'S_ZIP' columns.

```python
df_subset[['SITE_ADDR', 'S_CITY', 'S_ZIP', 'SALE_PRICE']].dropna(axis=0, subset=['S_CITY', 'S_ZIP'])
```

## 1.3. Filling nulls

We can fill null values using the <b>.fillna()</b> method.

Let's use the <b>.fillna()</b> method to fill missing values in the 'SALE_PRICE' field.

```python
df['SALE_PRICE']
```

The number -9999 is often used to represent null values in datasets where only integers are allowed.

Let's fill the null values with -9999.

```python
df['SALE_PRICE'].fillna(-9999)
```

The <b>.fillna()</b> method provides for a number of keyword arguments, allowing you to specify more sophisticated filling logic.

## 1.3. A note on null types

In regular python, nulls are usually represented by the "None" object. Try typing "None" into the shell (without the quotes) and hitting ENTER. Instead of throwing an undefined variable error, the interpreter should return "None".

For numeric values, pandas generally represents nulls as "np.nan". From the name, you can see that this object is taken from the numpy library (remember, we imported numpy as np). The "nan" in "np.nan" means "not a number". It is similar to the native python "None" object, except that it is typed as a 64-bit float. For more info, see the docs here:

http://pandas.pydata.org/pandas-docs/stable/gotchas.html#choice-of-na-representation

For strings, null values can be represented as an empty string: "" or ''.

# 2. Dealing with duplicates

## 2.1. Getting unique entries

Use the <b>.unique()</b> method to get unique values in a Series.

```python
df_subset['SITE_ADDR'].unique()
```

If there are duplicates, the length of the <b>.unique()</b> array will be less than the length of the Series.

```python
print len(df['SITE_ADDR'])
print len(df['SITE_ADDR'].unique())
```

## 2.2. Finding duplicates

The <b>.duplicated()</b> method returns a boolean array indicating duplicate values in a given Series or DataFrame.

```python
addresses = df_subset['SITE_ADDR'].dropna()

addresses.duplicated()
```

The following addresses occur more than once in the database.

```python
addresses.loc[addresses.duplicated()]
```

## 2.3. Dropping duplicates

We can drop unwanted duplicates using the <b>.drop_duplicates()</b> method.

Use the <b>subset</b> keyword argument to drop rows containing duplicates in a certain column or subset of columns.

In this case, we want to drop rows containing duplicates in the 'SITE_ADDR' column.

```python
df_subset.drop_duplicates(subset=['SITE_ADDR'])
```

By default, the <b>.drop_duplicates()</b> takes the first duplicate entry.

To take the last duplicate entry, use the <b>take_last</b> keyword argument.

```python
df_subset.drop_duplicates(subset=['SITE_ADDR'], take_last=True)
```
