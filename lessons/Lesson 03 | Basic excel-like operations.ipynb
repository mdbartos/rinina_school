## 2.2. Applying descriptive functions on DataFrames

DataFrames contain two axes: the index and the columns.

Thus, when finding the sum, min, max, or mean of a DataFrame, we need to specify which axis we're summarizing:

- <b>axis=0</b> will apply the operation across each row

- <b>axis=1</b> will apply the operation across each column

These operations will default to axis=0 if no argument is supplied.

```python
df.mean(axis=1) #Will give you the mean of each row (i.e. from left to right)
```

```python
df.mean(axis=0) #Will give you the mean of each column (i.e. from top to bottom)
```

# 3. Find and replace

### 3.1. Finding values in a DataFrame or Series

There are several ways to find values in a dataframe.

Let's use a simple example dataframe containing a list of animals and their attributes.

Note: the methods presented here are best used for searching for entire cell values. If you want to search for part of a string, take a look at the <b>str.contains()</b> method in <b>Lesson 12: String Methods, Pt. 1</b>

```python
animal_d = {
    'NAME' : ['bat', 'goldfish', 'dog', 'owl', 'cat'],
    'CLASS' : ['mammal', 'fish', 'mammal', 'bird', 'mammal'],
    'WEIGHT_LBS' : [0.5, 0.2, 50, 4, 6],
    'COLOR' : ['black', 'gold', 'brown', 'brown', 'gray'],
    'LENGTH_IN' : [6, 3, 36, 10, 14]
    }

animal_df = pd.DataFrame.from_dict(animal_d).set_index('NAME')

animal_df
```

Looking for values in a Series can generally be accomplished through boolean indexing.

```python
animal_df[animal_df['CLASS'] == 'mammal']
```

Alternatively, if we want to look for certain values over the entire DataFrame, we can use boolean indexing, or the <b>.where()</b> method to return a same-size array where all non-matching values are "masked" with nulls.

```python
animal_df[animal_df > 3]
```

```python
animal_df.where(animal_df == 6)
```

We can lookup the integer indices of a value using the <b>np.where()</b> method (not to be confused with the <b>pd.DataFrame.where()</b> method).

```python
coords = np.where(animal_df == 6)

coords
```

Or in (row, column) form:

```python
zip(*coords)
```

```python
animal_df.iloc[coords]
```

### 3.2. Replacing values

Values can be replaced using the <b>.replace()</b> method.

Note: this method is best used for replacing entire cell values. If you want to replace part of a string, take a look at the <b>str.replace()</b> method in <b>Lesson 12: String Methods, Pt. 1</b>

```python
animal_df.replace('gold', 'orange')
```

Pass a dict into the <b>.replace()</b> method to replace multiple values.

```python
animal_df['WEIGHT_LBS'].replace({6:8, 4:5})
```

# 4. Sorting data

### 4.1. Sorting a Series

Use the <b>.order()</b> method to sort a Series.

This method sorts from low to high by default.

```python
animal_df['WEIGHT_LBS'].order()
```

To sort a Series from high to low, enter the keyword argument <b>ascending=False</b>

```python
animal_df['WEIGHT_LBS'].order(ascending=False)
```

### 4.2. Sorting a DataFrame

To sort a dataframe by a particular column, enter the column name as an argument in the <b>.sort()</b> method.

```python
animal_df.sort('WEIGHT_LBS')
```

To sort from highest to lowest (or Z to A), set the keyword argument <b>ascending=False</b>.

```python
animal_df.sort('LENGTH_IN', ascending=False)
```

You can perform hierarchical sorts by entering a list of columns to sort by.

```python
animal_df.sort(['CLASS', 'COLOR'])
```

### 4.3. Sorting the index

To sort by index labels, use the <b>.sort_index()</b> method.

This method targets rows (<b>axis=0</b>) by default.

```python
animal_df.sort_index()
```

You can sort column labels using the <b>axis=1</b> keyword argument.

```python
animal_df.sort_index(axis=1, ascending=False)
```

## 5.1. Creating a line graph

The line graph is the easiest one to make. Simply call the <b>plot()</b> method on two arrays of the same size.

```python
plot(x,y)
```

In this case, the x data will be the index of the DataFrame (the year), and the y data will be the Maricopa county population.

### 5.1.1. Adding axis labels, titles, legends

```python
plot(df['Yavapai'].index, df['Yavapai'], label='Yavapai')
plot(df['Pinal'].index, df['Pinal'], label='Pinal')

xlabel('Year')
ylabel('Population')
title('County Populations, Yavapai and Pinal')
legend(loc=2)
```

## 5.3. Creating a scatter plot

We can create a scatter plot using the <b>scatter()</b> method.

As with the line graph, the scatter plot requires two arguments: the x data and the y data.

```python
scatter(x,y)
```

The "c" argument can be used to input a color. This can either be a single color, or an array with the same length as the data.

In this case, we use the year as the color argument.

```python
scatter(df['Maricopa'], df['Pima'], c=[int(i) for i in df.index])

xlabel('Maricopa County Population')
ylabel('Pima County Population')
title('Maricopa vs. Pima County Population')
colorbar()
```
