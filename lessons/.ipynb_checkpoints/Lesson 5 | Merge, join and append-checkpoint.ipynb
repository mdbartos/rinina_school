# Merge, join and append

So far, we've learned how to join two DataFrames that share a common index using the concat() method.

However, this doesn't give quite the same expressive power as an SQL-style join.

For instance, what if we had two dataframes that shared multiple fields, and we wanted to join on all of the shared fields? The concat() method joins on a single shared index, making multiple-field joins very awkward.

Luckily, pandas has a number of convenience methods that make relational database-style joins very easy.

## SQL-style joins with "merge"

From the pandas API:

```python
merge(left, right, how='left', on=None, left_on=None, right_on=None,
      left_index=False, right_index=False, sort=True,
      suffixes=('_x', '_y'), copy=True)
```

Here’s a description of what each argument is for:

- left: A DataFrame object


- right: Another DataFrame object


- on: Columns (names) to join on. Must be found in both the left and right DataFrame objects. If not passed and left_index and right_index are False, the intersection of the columns in the DataFrames will be inferred to be the join keys


- left_on: Columns from the left DataFrame to use as keys. Can either be column names or arrays with length equal to the length of the DataFrame


- right_on: Columns from the right DataFrame to use as keys. Can either be column names or arrays with length equal to the length of the DataFrame


- left_index: If True, use the index (row labels) from the left DataFrame as its join key(s). In the case of a DataFrame with a MultiIndex (hierarchical), the number of levels must match the number of join keys from the right DataFrame


- right_index: Same usage as left_index for the right DataFrame


- how: One of 'left', 'right', 'outer', 'inner'. Defaults to inner. See below for more detailed description of each method


- sort: Sort the result DataFrame by the join keys in lexicographical order. Defaults to True, setting to False will improve performance substantially in many cases


- suffixes: A tuple of string suffixes to apply to overlapping columns. Defaults to ('_x', '_y').


- copy: Always copy data (default True) from the passed DataFrame objects, even when reindexing is not necessary. Cannot be avoided in many cases but may improve performance / memory usage. The cases where copying can be avoided are somewhat pathological but this option is provided nonetheless.

## Merging on a single field

### Inner join

As with concat(), an "inner join" returns only the entries that match in both dataframes.

In other words, it returns an <b>intersection</b> of the join keys (in both LEFT <b>AND</b> RIGHT).

```python
merge_on_address_inner = pd.merge(df_master, df_target, on='SITE_ADDR', how='inner')[['SITE_ADDR', 'PARCEL_NUM', 'FID_1']]

print '\n', merge_on_address_inner.head(15)
print '\n'
print len(merge_on_address_inner), 'matches\n'
```

### Outer join

An outer join returns the <b>union</b> of the join keys.

In other words, it returns all the join keys, both in the left and right dataframes (either in LEFT <b>OR</b> in RIGHT).

```python
merge_on_address_outer = pd.merge(df_master, df_target, on='SITE_ADDR', how='outer')[['SITE_ADDR', 'PARCEL_NUM', 'FID_1']]

print '\n', merge_on_address_outer.head(15)
print '\n'
print len(merge_on_address_outer), 'matches\n'
```

Note the null values in the FID_1 column. These represent entries in df_master that were not found in df_target.

## Merging on multiple fields

```python
ignore_street_type = pd.merge(df_master, df_target, on=['SNUM', 'STREET', 'DIR'], suffixes=['_MASTER', '_TARGET'], how='inner')

print '\n', ignore_street_type.head(15)
print '\n'
print len(ignore_street_type), 'matches\n'
```

### What's with the suffix?

The suffix argument in our merge helps us to keep track of which fields are from which original dataframe. Fields with the suffix 'MASTER' are from master_df, while fields with the suffix 'TARGET' are from target_df.

Compare 'SITE_ADDR_MASTER' with 'SITE_ADDR_TARGET'. In row one, 'SITE ADDR MASTER' has the extension 'DR', while 'SITE_ADDR_TARGET' has the extension 'AVE'.

This is because we decided to ignore the street extension in our merge. When we specified the join fields for our merge, we included fields for street number, cardinal direction and street name ('SNUM', 'DIR', and 'STREET'), but we did not include the field for the extension ('STYPE').

Note that by ignoring the street extension in our merge, we increased the number of matches from 1664 to 2572 (at the expense of accuracy).

## Using "join" for a more SQL-style expression

The join() method is the same as merge(), except that it puts the left dataframe on the outside of the function, like so:

```python
df1.join(df2)
```

This is closer to the way joins are expressed in SQL. It allows the user to "chain together" as many joins as he or she wishes.

```python
df1.join(df2).join(df3).join(df4)
```

etc., etc...

## Gluing together datasets with "append"

Append is mostly useful for "gluing together" datasets that share a set of columns.

This can be very convenient when you have time series data located in seperate files (say, one file for 1990, another for 1991, etc.).

It's equivalent to using concat with axis=0, except that, like join(), it puts the left dataframe on the outside of the function:

```python
df1.append(df2)
```
