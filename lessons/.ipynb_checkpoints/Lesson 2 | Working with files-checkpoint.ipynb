# 2. Reading files in pandas

The pandas library has many built-in methods for reading files. csv and txt files are generally the easiest to work with, although pandas also supports more advanced structures like excel files, and high-performance data structures like HDF5.

## 2.1. Reading formatted text files

To import plain csv and txt files as DataFrames, we can use the read_csv() method.

The read_csv() method can be used with <b>any formatted text file</b>, not just .csv files.

Let's read some census data from my public dropbox folder

Note: the read_csv() method can be used with local as well as remote files.

```python
file_name = 'https://dl.dropboxusercontent.com/u/14284692/az_census.csv'

df = pd.read_csv(file_name)
```

Now call the dataframe object.

```python
df
```

Wow, that was easy!

But wait, what's with the "Unnamed: 0" column? This happened because the first column in the input file has no header. The read_csv() method assigned a new index column, rather than using the first column. We can change this behavior by specifying the index column we want to use when calling the read_csv() method.

```python
df = pd.read_csv(file_name, index_col=0)

df
```

### 2.1.1. Specifying delimiters and headers

The read_csv() method assumes that the file is comma-delimited. For space or tab delimited files, specify the 'sep' argument:

```python
file_name = 'https://dl.dropboxusercontent.com/u/14284692/baker.csv'

df = pd.read_csv(file_name, sep=' ') #Where ' ' represents a space

df = pd.read_csv(file_name, sep='\t') #Where '\t' represents tab.

df
```

In this case, the input file doesn't have any headers, so we can specify them manually using the 'names' argument:

```python
df = pd.read_csv(file_name, sep='\t', index_col=0, names=['station_code', 'station_name', 'latitude', 'longitude'])

df
```

### 2.1.2. Working with poorly formatted data

Sometimes you'll encounter files which cannot easily be converted into rectangular tables. This is common among government excel tables, which often contain "titles" and explanatory remarks before the start of the actual data.

```python
file_name = 'https://dl.dropboxusercontent.com/u/14284692/Mining%20Materials%20Received%2C%20United%20States%2C%202007.csv'

pd.read_csv(file_name).head(10)
```

Wow, what a mess. Clearly we need to do some extra work before we can work with this data.

From the table above, it appears that the actual data begins on dataframe row 5, with the header on dataframe row 3.

We can solve this problem by explicitly passing in the row that the header starts (in this case, file row 4).

```python
mining_data = pd.read_csv(file_name, header=4)

mining_data
```

This dataframe still has a number of problems, however.

- There are empty rows (containing only nulls).
- There are empty columns (containing only nulls).
- The last 10 rows are "footnotes" that were captured by the parser.

We can fix these problems using the dropna() method.

```python
#First, let's remove the footer by cutting off the last ten rows.

mining_data = mining_data.loc[:552]

#This will drop empty columns.

mining_data.dropna(axis=1, how='all', inplace=True)

#This will drop empty rows.

mining_data.dropna(axis=0, how='all', inplace=True)
```

Now our data should be nice and clean!

```python
mining_data.head(10)
```

You can check out more optional arguments by typing pd.read_csv? in the ipython shell, or by visiting the pandas API online.

## 2.2. Working with excel files

To read excel files, call the read_excel method:

```python
xl_file_name = 'https://dl.dropboxusercontent.com/u/14284692/Mining%20Materials%20Received%2C%20United%20States%2C%202007.xls'

xl_df = pd.read_excel(xl_file_name)
```

If you want to read a specific sheet, put it in the second argument of the function call. The read_excel() method will default to the first sheet.

```python
sheet_name = 'EC0721I3'

xl_df = pd.read_excel(xl_file_name, sheet_name)
```

```python
xl_df.head(5)
```

Keep in mind that excel files are often not formatted exactly like rectangular tables. You may need to pass in a skiprows argument to skip extraneous headers, etc.

```python
xl_df = pd.read_excel(xl_file_name, sheet_name, skiprows=4, skipfooter=10)

xl_df.dropna(axis=1, how='all', inplace=True)
xl_df.dropna(axis=0, how='all', inplace=True)
```

```python
xl_df.head(5)
```

Sometimes, you will want to drop rows that contain nulls only in a particular column or list of columns.

For instance, if we look at the footer of the original file, we'll see that the header and footer contain nulls in the third column ('Material or Fuel Code') column, while the actual data always lists a 'Material or Fuel code'.

Using this knowledge, we can drop the header and footer in one step by telling the read_csv() method to drop rows that contain a null in the third column (which will default to 'Unnamed: 3' unless column names are explicitly passed):

```python
xl_df = pd.read_excel(xl_file_name, sheet_name)

xl_df.dropna(subset=['Unnamed: 3'])
```

```python
xl_df = xl_df.dropna(subset=['Unnamed: 3']).dropna(axis=[0,1], how='all')
xl_df.columns = list(xl_df.loc[3])
xl_df = xl_df.drop(3)
```

# 3. Files and dicts

Okay. Now before we move on, let's start with a refresher on dictionaries.

Why, you ask?

Because when you start importing files into your python session, you're going to need a systematic way of keeping track of all your files, and their names.

For instance, say you want to(1) loop through a directory and import a bunch of csv files as DataFrames, and (2) you want the variable name of each DataFrame to correspond to the filename. This can't be done! There's no way to create new variable names on-the-fly in a for loop.

Instead, we use dictionaries. Here is how you would loop through a directory, and import all the tables to a dictionary (d):

```python
d = {}

for filename in os.listdir('.'):
    if filename.endswith('csv'):
        df = pd.read_csv(filename)
        d.update( { filename : df } )
```

The .update() method inserts a key:value pair into a dictionary. In this case, the key is the filename, and the value is the corresponding dataframe.

We can now access any imported table by typing d[_FILENAME_]

The following three methods are really important for dealing with dictionaries, esp. in for loops:

```python
d.keys() #Should give you a list of all the filenames of the imported tables
```

```python
d.values() #Will return all the tables you imported in a list.
```

```python
d.items() #Will enumerate the key:value pairs. This is useful in for loops.
```
