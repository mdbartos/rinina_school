The stack() method allows you to pivot the columns of a dataframe onto a level in the index.

The unstack() method is the inverse operation of stack. It takes a level from a hierarchical index and pivots it onto the DataFrame's columns.

# 1. Using Stack

First, let's import the modules and dataset.

In this lesson, we'll be working with data from the Bureau of Labor statistics.

The numbers in the following DataFrame represent the total labor force for the years 1993 to 2013 (on a monthly time step).

## 1.1. Using stack to linearize the data

Now let's say we want to create a time series of total labor force enrollment over time.

This is difficult to accomplish when each month has its own column.

To solve this problem, we can call the stack() method to create a hierarchical index:

```python
labor.stack()
```

## 1.2. Plotting the stacked data

```python
# Let's reset the hierarchical index to a single level, and rename the columns to something more meaningful.

stacked = labor.stack().reset_index().rename(columns={'level_1':'Month', 0:'Labor Force'})
stacked.head()
```

```python
# We can use the 'Year' and 'Month' columns to construct a date column.
# First, let's concatenate the 'Year' and 'Month' columns together. 
stacked['date'] = stacked['Year'].astype(str) + stacked['Month']

# Now we can use the strptime method to convert the 'date' column to a datetime type.
stacked['date'] = stacked['date'].apply(lambda x: datetime.datetime.strptime(x, '%Y%b'))

# Now let's see the result.
stacked.head()
```

# 2. Using unstack

Unstack is simply the inverse operation of stack.

At its simplest level, it turns a level of the hierarchical index into columns.

```python
stacked = labor.stack()

stacked.unstack()
```
